```python
import matplotlib.pyplot as plt
import numpy as np

x = [100, 500, 1000, 5000, 10000, 50000, 100000]
y1 = [0.00105817, 0.036778, 0.100455, 2.00656, 6.53678, 159.549, 632.609]
y2 = [0.000315084, 0.0150654, 0.00283504, 0.0159017, 0.0428772, 0.198829, 0.424171 ]
y3 = [0.000261792, 0.00120363, 0.00236733, 0.0126353, 0.0283997, 0.14042, 0.302209]

plt.figure()
plt.plot(x, y1, label='Пузырьковая сортировка')
plt.plot(x, y2, label='Сортировка кучей')
plt.plot(x, y3, label='Быстрая сортировка')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Размер массива')
plt.ylabel('Время')
plt.title('Сравнение времени работы сортировок')
plt.legend()
plt.show()
```
